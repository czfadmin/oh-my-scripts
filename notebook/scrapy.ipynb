```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
product_url="https://www.qualcomm.com/products"
results=[]
for i in range(1,10):
    response= requests.get(f"{product_url}/snapdragon-{i}-series-mobile-platforms")
    if response.ok:
        content= BeautifulSoup(response.content)
        tables= content.find_all("table")
        headers=[]
        product_result=[]
        info_result:list=[]
        for table in tables:
            th_list= table.select("thead>tr>th")
            for th in th_list:
                headers.append(th.text.strip())

        info_tr_list= tables[0].select("tbody>tr>td p")
        for info_tr in info_tr_list:
            product_result.append(info_tr.text.replace(' ','').replace('\n',' ').strip())

        info_tr_list2 = tables[1].select('tbody>tr')
        for info_tr in info_tr_list2:
            info_td_list:list=[]
            for info_td in info_tr.select('td'):
                info_td_list.append(info_td.text.replace(' ','').replace('\n',' ').strip())
            info_result.append(info_td_list)

        for j in range(len(product_result)):
            info_result[j].insert(0,product_result[j])

        product_df=pd.DataFrame(info_result,columns=pd.Series(headers))
        results.append(product_df)
result_df=pd.concat(results)
result_df.to_json('./data.json',orient="records")
print("ok")
```
